# [**60. 第k个排列**](https://leetcode-cn.com/problems/permutation-sequence/)
给出集合 [1,2,3,…,n]，其所有元素共有 n! 种排列。

按大小顺序列出所有排列情况，并一一标记，当 n = 3 时, 所有排列如下：

"123"
"132"
"213"
"231"
"312"
"321"
给定 n 和 k，返回第 k 个排列。

说明：

给定 n 的范围是 [1, 9]。
给定 k 的范围是[1,  n!]。
示例 1:

输入: n = 3, k = 3
输出: "213"
示例 2:

输入: n = 4, k = 9
输出: "2314"

```python
class Solution:
    def getPermutation(self, n: int, k: int) -> str:
        k -= 1
        l = [str(i) for i in range(1,n+1)]
        l1 = []
        for i in range(1, n):
            k, b = divmod(k, i)
            l1.append(b)
        l1.append(k)
        ans = ''
        for i in l1[::-1]:
            ans += l.pop(i)
        return ans

#         def fink(ni, k):
#             if ni == n:
#                 return l.pop(k)
#             else:
#                 k, b = divmod(k, ni)
#                 return fink(ni+1, k) + l.pop(b)
#         return fink(1, k)
        
```
