# [**257. 二叉树的所有路径**](https://leetcode-cn.com/problems/binary-tree-paths/)
给定一个二叉树，返回所有从根节点到叶子节点的路径。

说明: 叶子节点是指没有子节点的节点。

示例:

输入:

   1
 /   \
2     3
 \
  5

输出: ["1->2->5", "1->3"]

解释: 所有根节点到叶子节点的路径为: 1->2->5, 1->3

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def binaryTreePaths(self, root: TreeNode) -> List[str]:
        ans = []
        if not root:
            return ans
        def wrap(root, pre=[]):
            pre.append(root.val)
            left, right = root.left, root.right
            if left:
                wrap(left, pre)
                pre.pop()
            if right:
                wrap(right, pre)
                pre.pop()
            if right is None and left is None:
                ans.append('->'.join(map(str, pre)))
        wrap(root)
        return ans
```
