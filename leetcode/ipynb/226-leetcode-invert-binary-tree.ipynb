```python
from typing import List, Dict, Tuple# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def invertTree(self, root: TreeNode) -> TreeNode:
        def reverse(root):
            if root:
                root.right, root.left = root.left,root.right
                reverse(root.left)
                reverse(root.right)
        reverse(root)
        return root             
```
